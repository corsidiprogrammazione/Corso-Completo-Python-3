```python
import timeit
```

```python
"-".join(str(n) for n in range(0,100))
```

```python
timeit.timeit('"-".join(str(n) for n in range(0,100))',number=10000)
```

```python
timeit.timeit('"-".join([str(n) for n in range(0,100)])',number=10000)
```

```python
timeit.timeit('"-".join(map(str,range(100)))',number=10000)
```
